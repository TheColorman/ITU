**<span style="background: Yellow">4.4.15 - Yellow</span>**  What happens to Bellman–Ford if there is a negative cycle on the path from s to v and then you call `pathTo(v)`?  

*Solution*:  
It throws an error. The Bellman-Ford algorithm can detect negative cycles. If there is a negative cycle, the queue of vertices will *not* be empty after execution.  

**<span style="background: Yellow">4.4.31 - Yellow</span>**  Given a weighted line graph (undirected connected graph, all vertices of degree 2, except two endpoints which have degree 1), describe how you would devise an algorithm that preprocesses the graph in linear time and can return the distance of the shortest path between any two vertices in constant time.

*Solution*:  
Keep an array of cumulative distance. That is, if the edges in order have weight 2, 3, 4, the array would be `[2, 5, 9]`. Then, when wanting to get the distance between `a` and `b`, you do `distTo[b]-distTo[a]`.
