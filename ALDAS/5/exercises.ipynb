**<span style="background: Yellow">2.4.11 - Yellow</span>**  Suppose that your application will have a huge number of insert operations, but only a few *remove the* maximum operations. Which priority-queue implementation do you think would be most effective: heap, unordered array, or ordered array? Explain why.

*Solution*:  
Heap: Insert takes $1+\log_2n$. Remove takes $2\log_2n$.  
Unordered array: Insert takes $1$. Remove takes $n$.  
Ordered array: Insert takes $1+2n$.  Remove takes $1$.  
  
If we have so few remove operations that it has no effect on our time, only the insert operation is worth looking at. The one with the fastest insert operation is the unordered array.  

**<span style="background: Yellow">2.4.12 - Yellow</span>**  Suppose that your application will have a huge number of *find the maximum* operations, but a relatively small number of *insert* and *remove the maximum* operations. Which priority-queue implementation do you think would be most effective: heap, unordered array, or ordered array? Explain why.

*Solution*:  
The only time worth looking at is find maximum.  
Heap: $1$.  
Unordered: $n$.  
Ordered: $1$.  
So the heap and ordered array have the same time for find operations. In that case, we take the fastest one for other operations.  
Heap is $1+\log_2n$ for insert and $2\log_2n$ for remove.  
Ordered is $1+2n$ for insert and $1$ for remove.  
Heap is way faster for $n$ insert + $n$ remove operations, so we'll go with that.  

**<span style="background: Yellow">2.4.21 - Yellow</span>**  Explain how to use a priority queue to implement the stack and queue from week 2.  

*Solution*:  
It is possible by assigning a numeric identifier to each element, and storing the identifiers in the priority queue.  
Stack:  
Every time you add an element, add it to the end of an array. Add the index to the priority queue.  
When popping an element, find and remove the largest index stored in the p-queue and return the element in that position.  
  
Queue:  
Enqueue: Add an element to the end of an array. This index is the max. Add this index to the p-queue.  
Dequeue: Find the smallest index in the p-queue and return element at that position.
