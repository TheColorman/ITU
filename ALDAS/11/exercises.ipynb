**<span style="background: Yellow">3.3.22 - Yellow</span>**  Find a sequence of keys to insert into a BST and into a red-black BST such that the height of the BST is less than the height of the red-black BST, or prove that no such sequence is possible.  

*Solution*:  
Such a sequence is not possible.  
For any sequence of insertions, the height of the red-black BST will always increase slower than that of the BST, as the red-black BST compresses heights by rotating red edges.
