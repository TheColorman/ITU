**<span style="background: Yellow">1.4.29a - Yellow</span>**  This was a red exercise in week 3. The focus this week is on the argument
for the amortized number of stack operations.  
Steque with two stacks. Design a steque (a stack-ended queue or steque is a data type that supports `push`, `pop`, and `enqueue`) with two stacks so that each steque operation takes a constant amortized number of stack operations.  

*Solution*:  

````python
from itu.algs4.fundamentals.stack import Stack
class Stequeue:
    def __init__(self) -> None:
        self.main = Stack()
        self.secondary = Stack()
        self.size = 0
    
    def push(self, item):
        self.main.push(item)
        self.size += 1
    
    def enqueue(self, item):
        self.secondary.push(item)
        self.size += 1
    
    def pop(self):
        if self.main.is_empty():
            if self.secondary.is_empty():
                return None

            for _ in range(self.secondary.size()):
                self.main.push(self.secondary.pop())
        
        self.size -= 1
        return self.main.pop()
    
    def size(self):
        return self.size

stequeue = Stequeue()
stequeue.push(1) # [ 1 ]
stequeue.push(2) # [ 1, 2 ]
stequeue.push(3) # [ 1, 2, 3 ]
print(stequeue.pop()) # [ 1, 2 ]
print(stequeue.pop()) # [ 1 ]
print(stequeue.pop()) # [ ]
stequeue.push(1) # [ 1 ]
stequeue.push(2) # [ 1, 2 ]
stequeue.push(3) # [ 1, 2, 3 ]
stequeue.enqueue(4) # [ 4, 1, 2, 3 ]
stequeue.enqueue(5) # [5, 4, 1, 2, 3 ]
stequeue.push(6) # [ 5, 4, 1, 2, 3, 6]
print(stequeue.pop()) # [ 5, 4, 1, 2, 3 ]
print(stequeue.pop()) # [ 5, 4, 1, 2 ]
print(stequeue.pop()) # [ 5, 4, 1 ]
print(stequeue.pop()) # [ 5, 4 ]
print(stequeue.pop()) # [ 5 ]
print(stequeue.pop()) # [ ]
````

**<span style="background: Red">Set Union - Red</span>**  In the Set Union problem we have $n$ elements, that each are initially in $n$ singleton sets, and we want to support the following operations:  
- Union(A,B): Merge the two sets A and B into one new set $C=A\cup B$, destroying the old sets.  
- SameSet(x,y): Return `true`, if x and y are in the same set, and `false` otherwise.  
  
We can implement it the following way. Initially, give each set a color. When merging two sets, recolor the smallest one with the color of the larger one (break ties arbitrarily). To answer SameSet queries, check if the two elements have the same color.  
- a) Analyze the worst case cost of the two operations.  
- b)  Show that the amortized cost is $O(\log n)$ for Union and $O(1)$ for SameSet. That is, show that any sequence of $m$ unions and $s$ SameSet operations takes time $O(m \log n + s)$. Hint: Give a bound on the number of times an element can be recolored.  

*Solution*:  
To implement this we will have an array that is element-indexed and defines a color for every element.  
a)  
Worst case of a union operation will take $n/2$ if there are two equally sized sets. We need to update the color of every element in the other set.  
For SameSet it takes $O(1)$ as we just need to compare the values at `x` and `y`.  
b)  
If we continually union a sets until we only have one set, the final union will merge two sets and update $n/2$ elements. Before that, we will have 4 sets each with $n/4$ elements, unioned to two sets, updating $n/2$ elements. Again, before that we have 8 sets each with $n/8$ elements, updating half of the sets and $n/2$ elements.  
We do this until we reach $1$. So the sequence is:  
$1,2,4,\dots,\frac{n}{8},\frac{n}{4},\frac{n}{2}$. For each step in the sequence, we do $n/2$ operations. Since the sequence doubles for every element, it is the logarithm of $n$ with base 2. We therefore have $O(\log_2n)$.  
For the SameSet operation, the same argument as in a) applies.
